# Indicators

*  regarding inputs:
    * #of inputs
    * transfered value in (sum of all input values)
    * largest_input
    * smallest_input
* regarding outputs:
    * #of outputs
    * transfered value out (sum of all output values)
    * largest_output
    * smallest_output
* misc
    * #of transactions in my block
    * script_type
    * time - transaction time







```python
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN anz_inputs int
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN sum_in float
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN max_in float
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN min_in float


%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN anz_outputs int
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN sum_out float
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN max_out float
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN min_out float

%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN anz_trans_in_block int
%sql ALTER TABLE tmp_cluster_bitcoin ADD COLUMN script_type varchar
```

```python
%sql select * from tmp_cluster_bitcoin limit 10
```

```python
%sql UPDATE tmp_cluster_bitcoin as source SET anz_inputs = upd.anz_inputs , \
    sum_in = upd.sumv \
    ,max_in = upd.maxv \
    ,min_in = upd.minv \
    FROM \
( \
    SELECT trans.txid,count(1) as anz_inputs, sum(vin.value) as sumv, min(vin.value) as minv, max(vin.value) as maxv \
 from transaction as trans JOIN vin as vin \
    on trans.txid=vin.txid group by trans.txid \
) as upd where source.txid=upd.txid
```

```python
%sql UPDATE tmp_cluster_bitcoin as source SET anz_outputs = upd.anz_outputs , \
    sum_out = upd.sumv \
    ,max_out = upd.maxv \
    ,min_out = upd.minv \
    FROM \
( \
    SELECT trans.txid,count(1) as anz_outputs, sum(vout.value) as sumv, min(vout.value) as minv, max(vout.value) as maxv \
 from transaction as trans JOIN vout as vout \
    on trans.txid=vout.txid group by trans.txid \
) as upd where source.txid=upd.txid
```

```python
%sql UPDATE tmp_cluster_bitcoin as source SET anz_trans_in_block = upd.anz \
    FROM ( select bitcoin_block_hash, count(1)  as anz \
    FROM tmp_cluster_bitcoin group by bitcoin_block_hash ) as upd \
    where upd.bitcoin_block_hash = source.bitcoin_block_hash
```

```python
#script_type
%sql UPDATE tmp_cluster_bitcoin as source SET script_type = upd.script_type \
    FROM (  \
         SELECT txid,CASE WHEN count(distinct(type))>1 THEN 'mehr' ELSE max(type) END as script_type \
         FROM vout group by txid \
         ) as upd \
    where upd.txid = source.txid
```
